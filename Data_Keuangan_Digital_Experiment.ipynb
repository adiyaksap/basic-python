##**Insert Data**

```python
## Mengambil Data csv SPIP Desember dan import add ons
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn


df = pd.read_csv('/content/drive/MyDrive/Nyoba_Data/Rangkuman Data Digital Finance_Des.csv',sep = ';',index_col = 0)
display(df)

```

##**Mengetahui Jumlah Instrumen Dari Masing-Masing Alat Pembayaran**

```python
import datetime
from pandas_datareader.data import DataReader

plt.style.use('ggplot')

plt.subplot(121)
df['Jumlah Instrumen Kartu ATM per Unit'].plot(kind='line',figsize=(50,10),color='red',linewidth=3,marker='x')
plt.title('Jumlah Kartu ATM Beredar Di Indonesia per Unit Periode 2018 - 2023',size=30,weight='bold')
plt.xlabel('Periode',size=15)
plt.ylabel('Unit',size=15)
plt.legend()

plt.subplot(122)
df['Jumlah Instrumen Uang Elektronik per Unit'].plot(kind='line',figsize=(50,10),color='blue',linewidth=3,marker='x')
plt.title('Jumlah Uang Elektronik Beredar Di Indonesia per Unit Periode 2018 - 2023',size=30,weight='bold')
plt.xlabel('Periode',size=15)
plt.ylabel('Unit',size=15)
plt.legend()


plt.show()
plt.savefig('jumlahunit.jpg')
```

#**Mengetahui Volume Transaksi Dari Masing-Masing Alat Pembayaran**


```python
import datetime
from pandas_datareader.data import DataReader

plt.style.use('fivethirtyeight')

plt.subplot(121)
df['Volume Transaksi  Kartu ATM'].plot(kind='line',figsize=(50,10),color='red',linewidth=3,marker='x')
plt.title('Volume Transaksi Kartu ATM Di Indonesia Periode 2018 - 2023',size=30,weight='bold')
plt.xlabel('Periode',size=15)
plt.ylabel('Volume (Rp)',size=15)
plt.legend()

plt.subplot(122)
df['Volume Transaksi  Uang Elektronik'].plot(kind='line',figsize=(50,10),color='blue',linewidth=3,marker='x')
plt.title('Volume Transaksi Uang Elektronik Beredar Di Indonesia Periode 2018 - 2023',size=30,weight='bold')
plt.xlabel('Periode',size=15)
plt.ylabel('Volume (Rp)',size=15)
plt.legend()


plt.show()
plt.savefig('volumetransaksi.jpg')
```
